```python
# import dependancies 
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
plt.style.use('seaborn-deep')
```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        file = os.path.join(dirname, filename)
        print(file)
```

```python
#read dataset 
df = pd.read_csv(file)
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
# check if any values are 0. If there are missing values, would need to clean dataset 
df.isnull().sum(axis=0)
```

```python
df['Color'] = df['Color'].str.lower() #convert all colour values to lowercase 
df[['Type','Color']].value_counts().sort_values() #sort all colours type with how many of each colour there are, you can see if there are any colours/names that can be grouped together 
```

```python
df['Color'].replace('white-yellow','yellow-white', inplace=True)
df['Color'].replace('yellowish white','yellow-white', inplace=True)
df['Color'].replace('blue white','blue-white', inplace=True)
df_sorted = df[['Type','Color']].value_counts().sort_values()
df_sorted 
```

```python
df[['Spectral_Class','Type']].value_counts().sort_values()
```

```python
palette = ["#ff3333", "#80d4ff", "#d9d9d9", "#e6e600", "#ffcc00", "#0080ff", "w", "#ff6600", "#e6e600", "#ff9980"]
sns.catplot(x='Spectral_Class', y ='Temperature', hue='Color', kind='swarm', data = df, height = 7, aspect = 2, ci = None, palette = palette) # scatterplot
sns.set(rc={'axes.facecolor':'b', 'figure.facecolor':'b'})
plt.show()
```
